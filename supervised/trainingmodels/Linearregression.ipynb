```python
# undestanding the linear regression\
# generating the ranmdom data
import matplotlib.pyplot as plt

import numpy as np
x=2*np.random.rand(100,1)
y=4 + 3 * x + np.random.randn(100, 1)
fig, ax=plt.subplots()
ax.scatter(x,y)

ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_title("X vs Y")

plt.show()


```

```python
# predicting the best value of the theta 
X_b = np.c_[np.ones((100, 1)), x] # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
# making predictions using the  above value
x_new=np.array([[1],[2]])
x_new_b = np.c_[np.ones((2, 1)), x_new]
x_predict=x_new_b.dot(theta_best)
print(x_predict)
```

```python
plt.plot(x_new ,x_predict,'r--')
plt.plot(x, y, "b.")

plt.show()
```

```python
# performing the using the sckit learn '
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
print("printing the intercept and coefficients")
print(reg.intercept_)
print(reg.coef_)
```

```python
# predicting the new data
reg.predict(x_new)
```

```python
# underrstanding of optimization techniques like gradient descent ans Stochastic , batch and mini batch is done 
# implementing the stochastic regressior
from sklearn.linear_model import SGDRegressor
```

```python
reg_s=SGDRegressor(max_iter=1000, tol=1e-3, penalty=None, eta0=0.1)
```

```python
reg_s.fit(x,y.ravel())
```

```python
print(reg_s.predict(x_new))
```
